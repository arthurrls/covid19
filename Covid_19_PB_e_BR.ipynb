```
#@title
#@markdown - 
print("Por favor, aguarde.")
import urllib.request
import json
import requests

from pandas.io.json import json_normalize
import json

import pandas as pd

from google.colab import files
from google.colab import output

# colunas: data__records__city e data__records__Casos
ob1 = pd.read_excel("https://github.com/arthurrls/covid19/raw/master/cidades.xlsx")

url_ob = "https://superset.plataformatarget.com.br/superset/explore_json/?form_data=%7B%22slice_id%22%3A1549%7D&amp;csv=true"
request_ob = urllib.request.Request(url_ob)
response_ob = urllib.request.urlopen(request_ob)
#print (response_ob.read().decode('utf-8'))

r_ob = requests.get(url_ob)
file = r_ob.json()
df_ob = file['data']

################



ob2 = json_normalize(df_ob,'records')
ob2.to_csv('teste.csv')

dicionario_municipios = {"Brejo Cruz":"Brejo do Cruz", "São Miguel de Itaipú":"São Miguel de Taipu", "Pitimbú":"Pitimbu", "Santa Lúzia":"Santa Luzia"}

ob2["Município de Residência"]= ob2["Município de Residência"].replace(dicionario_municipios)
ob2.sort_values("Município de Residência", ascending=True)

ob3 = ob2.groupby(['Município de Residência']).size().to_frame('Óbitos').reset_index().sort_values("Município de Residência", ascending=True)
ob3.head()

ob1['cidade'] = ob1['cidade'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
ob1['cidade'] = ob1['cidade'].str.upper()

ob3['Município de Residência'] = ob3['Município de Residência'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
ob3['Município de Residência'] = ob3['Município de Residência'].str.upper()

ob4 = pd.merge(ob1["cidade"],ob3, left_on = 'cidade', right_on='Município de Residência', how = 'left')
del ob4['Município de Residência']

  ####

ob4["Óbitos"].fillna("0", inplace = True)
ob4['Óbitos'].apply(pd.to_numeric)
ob4["Óbitos"] = pd.to_numeric(ob4["Óbitos"])

ob4['Óbitos'].sum()

ob4.to_excel(r'/content/Covid_PB_obitos.xls', encoding='utf-8', index = False)
files.download('Covid_PB_obitos.xls')

output.clear()

soma_obitos = ob4['Óbitos'].sum()
print(f"O total de óbitos é de {soma_obitos}")
```
